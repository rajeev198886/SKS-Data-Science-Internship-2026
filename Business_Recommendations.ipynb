```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Telco_Customer_Churn_Dataset.csv (1).csv")

data.head()
```

```python
churn_rate = data["Churn"].value_counts(normalize=True) * 100

print(churn_rate)
```

```python
contract = pd.crosstab(
    data["Contract"],
    data["Churn"]
)

print(contract)

contract.plot(kind="bar", figsize=(8,5))

plt.title("Churn by Contract Type")

plt.ylabel("Customers")

plt.show()
```

```python
payment = pd.crosstab(
    data["PaymentMethod"],
    data["Churn"]
)

print(payment)

payment.plot(kind="bar", figsize=(10,5))

plt.title("Churn by Payment Method")

plt.ylabel("Customers")

plt.show()
```

```python
internet = pd.crosstab(
    data["InternetService"],
    data["Churn"]
)

print(internet)

internet.plot(kind="bar", figsize=(8,5))

plt.title("Churn by Internet Service")

plt.ylabel("Customers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Churn",
    y="MonthlyCharges",
    data=data
)

plt.title("Monthly Charges vs Churn")

plt.show()
```
