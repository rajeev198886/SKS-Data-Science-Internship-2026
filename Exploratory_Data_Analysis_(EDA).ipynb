```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
data = pd.read_csv("Telco_Customer_Churn_Dataset.csv.csv")

data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
churn_rate = data["Churn"].value_counts(normalize=True) * 100

print(churn_rate)
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Churn", data=data)

plt.title("Overall Customer Churn")
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="gender", data=data)

plt.title("Customer Distribution by Gender")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Partner", data=data)

plt.title("Partner Status")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(x="Dependents", data=data)

plt.title("Dependents Status")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(data["tenure"], bins=20)

plt.title("Tenure Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Churn", y="tenure", data=data)

plt.title("Tenure vs Churn")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="Contract", hue="Churn", data=data)

plt.title("Churn by Contract Type")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="PaymentMethod", hue="Churn", data=data)

plt.title("Churn by Payment Method")

plt.xticks(rotation=45)

plt.show()
```
